# 1.

```python
# Open the file... 

df = pd.read_csv("ted.csv", index_col = "title")    # with the 'index_col' keyword, the 'title' column is set as the index
```

```python
# ...and display basic information for each variable, the number of null values and type.

df.info()
```

```python
df.head(1)
```

# 2.

```python
# Each presentation is identified by a title.

# Assign a sequential numeric value to each title. Save the mapping in a dictionary...

index_dict = {}
for id in range(1,len(df.index) + 1):    # the 'id' variable goes from 1 to the total number of rows
    index_dict[df.index[id - 1]] = id    # the dictionary is filled with the pairs title : id

# ...and add a variable with this numeric identifier to the original dataframe.

df["title_id"] = index_dict    # the 'index_dict' dictionary is assigned to the new column 'title_id'
```

```python
df.head()
```

# 3.

```python
# One variable contains missing values. Print the possible values of this variable.

df["speaker_occupation"]
```

```python
df["speaker_occupation"].isnull().sum() # there are 6 null values in the 'speaker_occupation' column
```

```python
# Create a new dataframe containing only the rows with missing values in this variable...

df_null = df[df["speaker_occupation"].isnull()] # The isnull() method returns a boolean Series indicating missing values. 
                                                # 'df_null' contains rows from 'df' where "speaker_occupation" is null.
```

```python
df_null
```

```python
# ...and remove these rows from the original dataframe.

df.dropna(inplace = True) # The 'inplace=True' argument modifies the original DataFrame directly, dropping all rows with missing values. 
```

```python
df.info() # as requested now there are no 'null' quantities
```

# 4.

```python
# Generate descriptive statistics for the numerical variables present: Mean, median, mode, and standard deviation.

df_num_col = df.select_dtypes(include='number') # This selects all columns in df that have a numeric data type and creates a new DataFrame

stats = df_num_col.agg(['mean', 'median', 'std', lambda x: x.mode().iloc[0]]) # The agg() method is used to apply more aggregation functions 
                                                                              # to the DataFrame. The 'mean', the 'median', the 
                                                                              # 'standard deviation' and the 'mode' are calculated.
                                                                              # A lambda function is defined to calculate the mode of the column. 
                                                                              # Since .mode() can return multiple values (if there's a tie),
                                                                              # iloc[0] is used to select the first mode.


stats.rename({'<lambda>': 'mode'}, inplace = True) # This is used to rename '<lambda>' with 'mode' in the stats Dataframe
stats
```

```python
# Additionally, produce the correlation matrix for the numerical variables.

df_num_col.corr()
```

```python
# For each year, show the total number of presentations and the total number of views.

df["year"] = (pd.to_datetime(df["film_date"], unit = "s")).dt.year # Convert the "film_date" column from Unix timestamp (seconds) to datetime,
                                                                   # and extract the year into a new column called "year".

num_views_per_year = df.groupby("year")["views"].sum() # Group the DataFrame by the "year" column and calculate the total number of views 
                                                       # for each year. The result is stored in the variable "num_views_per_year".

num_pres_per_year = df.groupby("year")["year"].value_counts() # Group the DataFrame by the "year" column and count the number
                                                              # of presentations (entries) for each year. The result is stored
                                                              # in the variable "num_pres_per_year".

print(num_views_per_year)
print("\n")
print(num_pres_per_year)
```

```python
# Determine which year had the highest number of views per presentation.

(num_views_per_year/num_pres_per_year).sort_values(ascending = False).head(1) # The number of views per presentation are sorted
                                                                              # in descending order. head(1) takes the highest value
```

```python
# Exactly the same can be done with the published date but this is less significant.

df["pub_year"] = pd.to_datetime(df["published_date"], unit = "s").dt.year
num_views_per_pub_year = df.groupby("pub_year")["views"].sum()
num_pres_per_pub_year = df.groupby("pub_year")["pub_year"].value_counts()
(num_views_per_pub_year/num_pres_per_pub_year).sort_values(ascending = False).head(1)
```
