```python
import re

# Function to normalize game names
def normalize_name(name):
    # Convert to lowercase
    name = name.lower()
    # Remove trademarks (e.g., ™, ®)
    name = re.sub(r'[™®]', '', name)
    # Remove other symbols (e.g., punctuation)
    name = re.sub(r'[^\w\s]', '', name)
    # Remove all spaces
    name = name.replace(' ', '')
    return name
```

```python
# Optimized code
temp_dict = {}
iter = 0

# Group the dataframe by 'normalized_title'
grouped = df3.groupby('normalized_title')

for normalized_title, group in grouped:
    temp_dict[normalized_title] = {}
    for _, row in group.iterrows():
        console = row['console']
        temp_dict[normalized_title][console] = {
            'critic_score': row['critic_score'],
            'total_sales': row['total_sales'],
            'na_sales': row['na_sales'],
            'jp_sales': row['jp_sales'],
            'pal_sales': row['pal_sales'],
            'other_sales': row['other_sales'],
            'release_date': row['release_date'],
            'last_update': row['last_update']
        }
    iter += 1
    print('Completion percentage: ', round(iter / len(grouped) * 100, 2), '%', end='\r')
```

```python
import requests
from concurrent.futures import ThreadPoolExecutor
import time

def get_steamspy_details(appid):
    """
    Fetch game details from SteamSpy API with throttling.
    """
    steamspy_url = f"https://steamspy.com/api.php?request=appdetails&appid={appid}"
    try:
        time.sleep(0.5)  # Delay each request to avoid hitting the server too fast
        response = requests.get(steamspy_url, timeout=10)
        if response.status_code == 200:
            return response.json()
    except Exception as e:
        print(f"Error for appid {appid}: {e}")
    return {field: np.nan for field in fields}  # Return NaNs for missing fields


# Define the fields to extract
fields = [
    'positive', 'negative', 'owners', 'average_forever', 'median_forever',
    'price', 'initialprice', 'discount', 'ccu', 'languages', 'genre', 'tags'
]

# Setup
data = {field: [] for field in fields}
appids = df2['appid'].tolist()
chunk_size = 200
total_appids = len(appids)
total_chunks = (total_appids + chunk_size - 1) // chunk_size  # rounds up
processed = 0  # running total of processed appids

for chunk_index, start in enumerate(range(0, total_appids, chunk_size), start=1):
    end = min(start + chunk_size, total_appids)
    chunk = appids[start:end]

    print(f"Chunk {chunk_index}/{total_chunks} | AppIDs {start}-{end} | Processed: {processed}/{total_appids}", end='\r')

    with ThreadPoolExecutor(max_workers=10) as executor:
        steamspy_data_list = list(executor.map(get_steamspy_details, chunk))

    for steamspy_data in steamspy_data_list:
        for field in fields:
            data[field].append(steamspy_data.get(field, np.nan))

    processed += len(chunk)

    time.sleep(10)  # throttle between chunks

# Add new columns to df2
for field in fields:
    df2[f'new_{field}'] = data[field]

# Save the updated dataframe
df2.to_csv('steam_games_enhanced_2.csv', index=False)
```
